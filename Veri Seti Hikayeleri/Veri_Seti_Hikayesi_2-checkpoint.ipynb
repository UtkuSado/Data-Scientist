```python
import seaborn as sns

tips = sns.load_dataset("tips")
df = tips.copy()
df.head()

# Veri Seti Hikayesi

# total_bill --> yemeğin toplam fiyati (bahşiş, vergi dahil)
# tip --> bahşiş
# sex --> ücreti ödeyen kişinin cinsiyeti(male= 0, female=1)
# smoker --> Grupta sigara içen var mı? (no= 0, yes=1)
# day --> gün (thur= 3, fri=4, sat= 5, sun= 6)
# time --> Ne zaman? (day= 0, night= 1)
# size --> Grupta kaç kişi var? 

#ÖNCE VERİ SETİNİ TANI!!
```

```python
# Kutu Grafiğinin Oluşturulması
# boxplot 

sns.boxplot(x= df["total_bill"]);

# Ortadaki çizgi medyanı ifade eder.
# Kutunun "Uzun Köşeleri" çeyreklik değerleri ifade ediyor.
```

```python
sns.boxplot(x= df["total_bill"], orient= "v");
```

```python
# Soru: Hangi günler daha fazla kazanıyoruz?

sns.boxplot(x= "day", y="total_bill", data= df);
```

```python
# Pazar günü gelen müşteri sayısı azken kazanılan para fazla, Cumartesi günü gelen müşteri sayısı çokken kazanılan para Pazara göre az.
```

```python
# Soru: Günün hangi saatinde daha fazla kazanıyoruz?

sns.boxplot(x= "time", y= "total_bill", data= df);
```

```python
# Akşam yemekleri de daha fazla para kazanılıyor. 
# En çok akşam saatleri ve pazar günleri çok kazanılıyor. !!!!!
```

```python
# Kişi sayısı kazancı...

sns.boxplot(x= "size", y= "total_bill", data= df);
```

```python
sns.boxplot(x= "day", y= "total_bill", hue= "sex", data= df);
```

```python
# Violin Grafiği --> Dağılım anlamda bilgi sunar.

sns.catplot(y= "total_bill", kind= "violin", data= df);
```

```python
# Violin Grafiği Çaprazlamalar

sns.catplot(x= "day", y= "total_bill", kind= "violin", data= df);
```

```python
sns.catplot(x= "day", y= "total_bill", hue= "sex", kind= "violin", data= df);
```

```python
# Korelasyon Grafiğinin Oluşturulması 

# Korelasyon --> Değişkenler arasındaki ilişkiyi ifade eden bir terimdir.
# Scatterplot --> İki değer arasındaki ilişkiyi ifade etmek için kullanılır.(sayisal değişkenlerin).

sns.scatterplot(x= "total_bill", y= "tip", data= df);
```

```python
# Korelasyon Çaprazlamalar

sns.scatterplot(x= "total_bill", y= "tip",hue= "time", data= df);
```

```python
sns.scatterplot(x= "total_bill", y= "tip",hue= "time", style= "time", data= df);
```

```python
sns.scatterplot(x= "total_bill", y= "tip",hue= "day", style= "time", data= df);
```

```python
# Sürekli değişkeni boyut olarak eklemek istersek...

sns.scatterplot(x= "total_bill", y= "tip", size= "size", data= df);
```

```python
sns.scatterplot(x= "total_bill", y= "tip", hue= "size", size= "size", data= df);
```

```python
# Doğrusal İlişkinin Gösterilmesi

import matplotlib.pyplot as plt #!!!!!!!!!!!!!!!!! Sadece pyplot fonksiyonun import ettik.

sns.lmplot(x= "total_bill", y= "tip", data= df);
```

```python
# Doğrunun etrafındaki açık renkli maviler --> sapmaları ifade eder.
```

```python
sns.lmplot(x= "total_bill", y= "tip", hue= "smoker", data= df);
```

```python
# Doğrunun eğimi yapacak olduğumuz yorumları belirler.
# Örneğin dogru turuncu dogru gibi ise --> Ödenen hesap arttıkça bahşişler artmıştır(Sıgara içmeyenler için.). Sıgara içenler için bu artışın bu kadar fazla 
# olmadığını eğimden anlarız.
```

```python
sns.lmplot(x= "total_bill", y= "tip", hue= "smoker", col= "time", data= df);
```

```python
# Makine öğrenmeden önce bizim ögrenmemiz gerekiyor! Arka taraftaki ilişkileri bilmemiz gerekiyor.
```

```python
sns.lmplot(x= "total_bill", y= "tip", hue= "smoker", col= "time", row= "sex", data= df);
```

```python
# Scatter Plot Matrisi --> Sayısal değişkenlerin hepsini matris olarak ifade etmek.

import seaborn as sns

iris = sns.load_dataset("iris")
df = iris.copy()
df.head()

# Çiçek türleri ve bunların özelliklerini ifade eden bir Veri Seti.
```

```python
df.dtypes
```

```python
df.shape #150 gözlemden, 5 değişkenden oluşuyor.
```

```python
sns.pairplot(df);
```

```python
# Baktığımızda veri setinde yer alan 4 değişkenin birbirleri arasındaki ilişki.
# Toz bulutu şeklinde bir grafik varsa bu, inceledğimiz 2 değişken arasında bir ilişki OLMADIĞINI gösterir.
# Oluşturulan scatterplot içerisinde gözlem noktaları birbirinden farklı köşelerde kümeleniyorsa çaprazlama durumundan kaynaklanan alt gruplar vardır.
```

```python
sns.pairplot(df, hue= "species");
```

```python
# Çeşitli noktalarda kümelenmeler varsa bu farklı yapılanmaları ifade eder.
```

```python
sns.pairplot(df, hue= "species", markers= ["o", "s", "D"]);

# markers --> birinci türü şu türden yap ikinci türü şu türden yap.
```

```python
sns.pairplot(df, kind= "reg");
```

```python
sns.pairplot(df, kind= "reg", hue= "species");
```
