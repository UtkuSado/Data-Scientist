```python
import seaborn as sns

fmri = sns.load_dataset("fmri")
df = fmri.copy()
df.head()

# Veri Seti Hikayesi >
# Beyine baglanan bir cihaz aracılıgıyla toplanan sinyalleri ifade eden bir veri seti
# subject --> Verilerin toplandıgı kişileri ifade ediyor.
# timepoint --> Zaman noktalarını ifade ediyor.
# event --> Birbirinden farklı olayları ifade ediyor.
# region --> Sinyalin toplandığı bölgeyi ifade ediyor.
# signal --> Gelen sinyali ifade ediyor.
```

```python
df.shape # 1064 adet gözlem birimi, 5 adet değişken.
```

```python
df["timepoint"].describe()
```

```python
df["signal"].describe()
```

```python
df.groupby("timepoint")["signal"].count() # Her bir sayma noktasında eşit sinyal alınmıştır.
```

```python
df.groupby("signal").count() # Her biri tekil olmuştur.
```

```python
df.groupby("timepoint")["signal"].describe()

# Birbirinden farklılaşan standart sapma var. Demekki ortalama etrafındaki durumlar birbirlerine göre değişebiliyor. Demekki bu değişimi oluşturan
# yapılar var. Standart sapmalar küçük birbirine yakın gibi ama biraz odaklanınca bagzı değerlerin birbiridnenoldukça farklılaştıgı görülüyor.
```

```python
# Çizgi Grafik ve Çaprazlamalar

sns.lineplot(x= "timepoint", y= "signal", data= df);

# Mavi çizgi --> ortalama değere karşılık geliyor.
# Etraftaki Mavi Alan --> Güven aralığı, standart sapma.
```

```python
sns.lineplot(x= "timepoint", y= "signal", hue= "event", data= df);
```

```python
sns.lineplot(x= "timepoint", y= "signal", hue= "event", style= "event", data= df);
```

```python
sns.lineplot(x= "timepoint", 
             y= "signal", 
             hue= "event", 
             style= "event",
             markers= True,
             dashes= False,
             data= df);

# Ortalama değerleri işaretleyip bize bir görsel sunmuş olacak.
```

```python
sns.lineplot(x= "timepoint", 
             y= "signal", 
             hue= "region", 
             style= "event",
             data= df);
```
